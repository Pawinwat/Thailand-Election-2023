# Data Preparation

```python
parties_df= pd.DataFrame(party_data)
```

```python
def highlight_party(s):
    ind = pd.Series(data=False, index=s.index)
#     print(s['color'])
    color = s['color']
    return [f'background-color: {color}'  for v in ind]

```

```python
parties_df.style.apply(highlight_party,  axis=1)
```

```python
buffer = []
for province in province_data:
    for party in province['result_party']:
        buffer.append([
            province['prov_id'],
            party['party_id'],
            party['party_cons_votes'],
            party['party_cons_votes_percent'],
            party['party_list_vote'],
            party['party_list_vote_percent']

        ])
```

```python
columns= ['prov_id','party_id','party_cons_votes','party_cons_votes_percent','party_list_vote','party_list_vote_percent']
vote_df = pd.DataFrame(buffer,columns=columns)
```

```python
vote_df
```

```python
parties_df['id'] = parties_df['id'].astype(int)
```

```python
full_df = vote_df.merge(parties_df,left_on='party_id',right_on='id')
full_df = full_df.merge(privince_df,left_on='prov_id',right_on='prov_id')

#privince_df
```

# Data Visualization

### Party-list vs Constituency By Province

```python
import plotly.graph_objects as go
provinces = full_df['province'].unique()

for province in provinces: 
    pov_df = full_df[full_df['province']==province]
    X=pov_df['province']

    fig = go.Figure(data=[
        go.Bar(name='Constituency', y=pov_df['party_cons_votes_percent'],x=pov_df['name'],marker_color=pov_df['color']),
        go.Bar(name='Party-list', y=pov_df['party_list_vote_percent'],x=pov_df['name'],marker_color=pov_df['color']),
    ])
    # Change the bar mode
    fig.update_layout(barmode='group',title=province)
    fig.show()
```

Move Forward and Pheu Thai are receive Party-list votes more than Constituency

### We found some pattern from result between Constituency and Party-list 

```python
full_df['df_party_con_diff'] = full_df['party_cons_votes_percent'] - full_df['party_list_vote_percent']
```

```python

fig = go.Figure(data=[
    go.Bar(name='diff', x=full_df['df_party_con_diff'],marker_color=full_df['color'],orientation='h')
])
fig.update_layout(
    title='party-con diff',
        width=1000,
    height=2000,
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=full_df['party_cons_votes_percent'],
        y=full_df['party_list_vote_percent'],
        mode='markers',
        marker_color=full_df['color'],
        text =full_df['name'],
        name='scatter'
            )
        )
fig.update_layout( xaxis_title='party_cons_votes_percent',
                   yaxis_title='party_list_vote_percent')
```

### Add trend lines

```python
from scipy import stats
def gen_point(slope,x,intercept):
    return x,slope * x + intercept

```

```python
line_buffer = []

for party in full_df['name'].unique():
    temp_party_df = full_df[full_df['name']==party]
    slope, intercept, r, p, std_err = stats.linregress(temp_party_df['party_cons_votes_percent'], temp_party_df['party_list_vote_percent'])
#     line_buffer.append({'name':party,'model':r})
    
    start = gen_point(slope,temp_party_df['party_cons_votes_percent'].min(),intercept)
    end = gen_point(slope,temp_party_df['party_cons_votes_percent'].max(),intercept)
    sub_fig = go.Scatter(x=[start[0],end[0]], y=[start[1],end[1]],name=party,marker_color=temp_party_df['color'].to_list()[0])
    line_buffer.append(fig)
    fig.add_trace(
        sub_fig
    )
```
